```python
#jangan lupa import lib
import pandas as pd
```

```python
#baca datanya yuk
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/iris_csv.csv')
data.head()
```

## 1. Mengurutkan dalam Pandas

```python
#Pandas bisa melakukan fungsi pengurutan sama halnya dengan ORDER BY dalam SQL

data.sort_values(by = 'iris_class', ascending = False)
```

```python
#bisa lebih dari satu kondisi

data.sort_values(by = ['iris_class', 'sepallength'], ascending = [False, True])
```

## 2. Filtering dalam Pandas

```python
#sama halnya SQL, dalam Pandas juga mengenang filtering yang sama seperti fungsi WHERE dan memilih kolom seperti SELECT

#memilih hanya beberapa kolom saja yang ingin ditampilkan
data[['iris_class', 'petallength']].head(10)
```

```python
#melakukan filter berdasarkan loc (string and condition)
data.loc[(data['iris_class'] == 'Iris-virginica')].head()
```

```python
#melakukan filter berdasarkan loc (string and condition) lebih dari satu condition
data.loc[(data.iris_class == 'Iris-virginica') & (data.petallength >= 1.4)].head()
```

```python
#Gabungan filter where dengan pemilihan kolom yang ditampilkan

data.loc[(data.iris_class == 'Iris-virginica') & (data.petallength >= 1.4), ['iris_class','petallength','petalwidth']].head()
```

```python
#Mengubah nilai kolom sesuai kondisi (Update set ... where ...)

data.loc[(data.iris_class == 'Iris-virginica') & (data.petallength >= 6.0) , ['iris_class']] = 'Special Flower'
```

## 3. Membuat dan Menambahkan kolom baru dari Dataframe

```python
data['sepallength_super'] = data['sepallength'] * 100
data.head(10)
```

## 4. Grouping DataFrame

```python
#dalam pandas juga mengenal istilah Grouping dan Agregasi sama halnya
#SQl yaitu fungsi GROUP BY

data.groupby('iris_class')
```

```python
#menghitung
data.groupby('iris_class').count()
```

```python
#menjumlahkan
data.groupby('iris_class').sum()
```

```python
#rata-rata
data.groupby('iris_class').mean()
```

```python
#minimal
data.groupby('iris_class').min()
```

```python
#mengembalikan dalam bentuk series

data.groupby('iris_class')['sepallength'].mean()
```

```python
#tapi kalo select kolom banyak, akan kembali jadi dataframe

data.groupby('iris_class')['sepallength','petalwidth'].mean()
```

```python
#Agregasi, Group by dan mengubah nama kolom

data.groupby(['iris_class', 'sepallength']).\
agg(average_petal = ('petallength','mean'))
```

```python
#Kombinasikan
"""
SELECT iris_class,
min(sepallength) as min_lgt,
min(sepalwidth) as min_wdt
FROM data
WHERE
sepallength >= 5
ORDER BY
iris_class DESC
"""
```

```python
data[['iris_class', 'sepallength','sepalwidth']].\
    loc[data['sepallength']>=5].\
    groupby(['iris_class']).\
    agg(min_lgt = ('sepallength','min'),min_wdt = ('sepalwidth','min')).\
    sort_values(by='iris_class', ascending = False)
```

## 5. Merge Part 1

```python
#Dalam Pandas ada juga fungsi untuk Inner, Full, Left, dan Right Join
#Baca Datanya dulu ya

t1 = pd.read_excel('merge.xlsx', sheet_name = 'Sheet1')
t2 = pd.read_excel('merge.xlsx', sheet_name = 'Sheet2')
```

```python
pd.merge(t1,t2, on = ['customer','item'])
```

```python
pd.merge(t1,t2, on = ['customer','item'], how='outer')
```

```python
pd.merge(t1,t2, on = ['customer','item'], how='left')
```

```python
pd.merge(t1,t2, on = ['customer','item'], how='right')
```
